```python
na_cols = train_df.isnull().sum().sort_values(ascending=False).loc[lambda s: s.ne(0)]
na_cols_list = na_cols.index.tolist()
na_num_cols_list = ['LotFrontage', 'GarageYrBlt', 'MasVnrArea']
na_obj_cols_list = [ele for ele in na_cols_list if ele not in na_num_cols_list]        
```

```python
print(na_num_cols_list)
```

```python
print(na_obj_cols_list)
```

```python
for col in na_obj_cols_list:
    unique_values = train_df[col].value_counts(dropna=False)
    null_count = train_df[col].isnull().sum()
    print("Unique Categories and their Frequencies:")
    print(unique_values)
    print(f"Null Values Frequency: {null_count}")
    print("="*50)
```
